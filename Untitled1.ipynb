```python
import pandas as pd

print("Libraries imported successfully")
```

```python
df= pd.read_csv("student_performance.csv")
print(df)
```

```python
print("Total rows:",df.shape[0])
print("Total columns:",df.shape[1])

print("Departments:",df['department'].unique())

print("Number of students in each department:")
print(df['department'].value_counts())
```

```python
highest_score = df['math_score'].max()
print("Highest score in math:", highest_score)

lowest_score = df['math_score'].min()
print("Lowest score in math:", lowest_score)

average_score = df['math_score'].mean().round(3)
print("Average score in math:", average_score)


```

```python
print("\n[4] gender-wise average scores")
print("-"*40)
gender_scores = df.groupby('gender')[['math_score','programming_score']].mean().round(3)
print(gender_scores)
```

```python
print("\n[5] department-wise average scores")
print("-"*40)
department_scores = df.groupby('department')[['math_score','programming_score']].mean().round(3)
print(department_scores)
```
