```python
total = df['math_score']+df['programming_score']+df['english_score']+df['science_score']
mean= total/4
mean_above_70 = mean[mean>70]
print(mean_above_70)
```

```python
cs_students=df[df['department']== 'Computer Science']
total_math_score = cs_students['math_score'].sum()
avg = total_math_score/13
print(avg)

elec_students=df[df['department']== 'Electronics']
total_math_score = elec_students['math_score'].sum()
avg = total_math_score/6
print(avg)

mech_students=df[df['department']== 'Mechanical']
total_math_score = mech_students['math_score'].sum()
avg = total_math_score/6
print(avg)

civil_students=df[df['department']== 'Civil']
total_math_score = civil_students['math_score'].sum()
avg = total_math_score/5
print(avg)
```
